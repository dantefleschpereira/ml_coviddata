## MergeData
Essa seção faz a união entre a tabela **Principal (Atendimentos)** e a tabela **Estudo**.
Estas tabelas vêm diretamente do hospital, sem tratamento.

---
A tabela **Principal** deve ter 41 colunas <br /> 
A Tabela **Estudo** deve ter 70 colunas

**NOTA 1**: O campo 'CD_PACIENTE' na tabela **Estudo** deve ser alterado para 'CD_PACIENTE_FROM_ESTUDO', apenas para identificar, na hora do merge, de onde o campor esta vindo.

#### Merge(Principal, Estudo)

Verificando se há valores faltantes no **CD_PACIENTE**:

```python
df_principal['CD_PACIENTE'].isnull().sum()
```

```python
_aux = []
cd_dos_pacientes = df_principal['CD_PACIENTE'].unique().tolist()

new_csv_file = open('./DataSteps/STEP_1_Principal_Estudo.csv', 'w+', encoding='utf8')
new_csv_file.write('\t'.join(df_principal.columns.to_list())+'\t'+'\t'.join(df_estudo.columns.to_list())+'\n')

for cd_p in cd_dos_pacientes:
    list_principal_values = df_principal.loc[df_principal['CD_PACIENTE']==cd_p].values.tolist()
    list_estudo_values = df_estudo.loc[df_estudo['CD_PACIENTE_FROM_ESTUDO']==cd_p].values.tolist()
    
    if len(list_principal_values) == 1:
        if len(list_estudo_values) == 1:
            row_values_principal = list_principal_values[0]
            row_values_estudo = list_estudo_values[0]
            for v in row_values_principal:
                _aux.append(str(v))
            for v in row_values_estudo:
                _aux.append(str(v))
            new_csv_file.write('\t'.join(_aux)+'\n')
            _aux = []
            
        if len(list_estudo_values) == 0:
            row_values_principal = list_principal_values[0]
            for v in row_values_principal:
                _aux.append(str(v))
            an_array = np.empty((1,70))
            an_array[:] = np.NaN
            for v in an_array.tolist()[0]:
                _aux.append(str(v))
            new_csv_file.write('\t'.join(_aux)+'\n')
            _aux = []

    if len(list_principal_values) > 1:
        if len(list_estudo_values) == 1:
            row_values_estudo = list_estudo_values[0]
            for row_value_principal in list_principal_values:
                for v in row_value_principal:
                    _aux.append(str(v))
                for v in row_values_estudo:
                    _aux.append(str(v))
                new_csv_file.write('\t'.join(_aux)+'\n')
                _aux = []
 
        if len(list_estudo_values) == 0:
            for row_value_principal in list_principal_values:
                for v in row_value_principal:
                    _aux.append(str(v))
                an_array = np.empty((1,70))
                an_array[:] = np.NaN
                for v in an_array.tolist()[0]:
                    _aux.append(str(v))
                new_csv_file.write('\t'.join(_aux)+'\n')
                _aux = []
        
new_csv_file.close()
```

## MergeData
Essa seção faz a união dos campos **DATA** e **HORA** da tabela **STEP_2_Principal_Estudo_UTI**. Essa união gera duas novas colunas: **DTHR_UTI_IN** e **DTHR_UTI_OUT**

---
**DTHR_UTI_IN** = DT_ENTRADA + ' ' + HR_ENTRADA <br />
**DTHR_UTI_OUT** = DT_SAIDA + ' ' + HR_SAIDA

```python
def __getDataIn(data_in):
    __d_i = parse(data_in, dayfirst=True)
    return str(__d_i.date().strftime('%d/%m/%Y'))
```

```python
def __getDataOut(data_out):
    __d_o = parse(data_out, dayfirst=True)
    return str(__d_o.date().strftime('%d/%m/%Y'))
```

```python
def __getHoraIn(hora_in):
    __h_i = parse(hora_in, dayfirst=True)
    return str(__h_i.time())
```

```python
def __getHoraOut(hora_out):
    __h_o = parse(hora_out, dayfirst=True)
    return str(__h_o.time())
```

```python
from dateutil.parser import parse
```

```python
pos_data_in = 117
pos_data_out = 119
pos_hora_in = 118
pos_hora_out = 120

new_csv_file = open('./DataSteps/STEP_3_Principal_Estudo_UTIcomDTHR.csv', 'w+', encoding='utf8')
new_csv_file.write('\t'.join(df_principal_estudo_uti.columns.to_list())+'\t'+'DTHR_UTI_IN'+'\t'+'DTHR_UTI_OUT'+'\n')

for row in df_principal_estudo_uti.values.tolist():
    if str(row[pos_data_in]) != 'nan':
        data_entrada = __getDataIn(row[pos_data_in])
    else:
        data_entrada = 'nan'
    
    if str(row[pos_data_out]) != 'nan':
        data_saida = __getDataOut(row[pos_data_out])
    else:
        data_saida = 'nan'
        
    if str(row[pos_hora_in]) != 'nan':
        hora_entrada = __getHoraIn(row[pos_hora_in])
    else:
        hora_entrada = ''
    
    if str(row[pos_hora_out]) != 'nan':
        hora_saida = __getHoraOut(row[pos_hora_out])
    else:
        hora_saida = ''
    
    data_hora_entrada = data_entrada+' '+hora_entrada
    data_hora_entrada = data_hora_entrada.strip()
    
    data_hora_saida = data_saida+' '+hora_saida
    data_hora_saida = data_hora_saida.strip()
    
    _aux = [str(value) for value in row]
    new_csv_file.write('\t'.join(_aux)+'\t'+data_hora_entrada+'\t'+data_hora_saida+'\n')
new_csv_file.close()
```

```python
def __getDifferenceDates(d_0, d_f):
    
    __d_0 = parse(d_0, dayfirst=True)
    __d_0 = __d_0.date()
    __d_f = parse(d_f, dayfirst=True)
    __d_f = __d_f.date()
    __diff = __d_f - __d_0
    
    return __diff.days
```

```python
def __getDifferenceInSeconds(d_0, d_f):
    #print(d_f)
    __d_0 = parse(d_0, dayfirst=True).ctime()
    __d_0 = parse(__d_0, dayfirst=True)
    __d_f = parse(d_f, dayfirst=True).ctime()
    __d_f = parse(__d_f, dayfirst=True)
    __diff = __d_f - __d_0
    
    return __d_0, __d_f, __diff.total_seconds()
```

```python
def __checkDatesColetaAlta(l_d0, l_df):
    if len(set(l_d0)) == 1 and len(set(l_df)) == 1:
        return 'equal'
    else:
        print('Datas Múltiplas: Coleta e Alta ')
```

```python
def __getListWthMultipleValues(l, cmb_prev_otr, cd_und, dts_in, dts_out):
    __aux = []
    comorb_prev_outras = '|'.join(list(set(cmb_prev_otr)))
    print(comorb_prev_outras)
    dates_in  = '|'.join(dts_in)
    print(dates_in)
    dates_out  = '|'.join(dts_out)
    print(dates_out)
    cds_unds = '|'.join(cd_und)
    print(cds_unds)
    for value in l:
        if l.index(value) == 84:
            __aux.append(comorb_prev_outras)
        elif l.index(value) == 115:
            __aux.append(cds_unds)
        elif l.index(value) == 121:
            __aux.append(dates_in)
        elif l.index(value) == 122:
            __aux.append(dates_out)
        else:
            __aux.append(str(value))
    return __aux
```

```python
def __isNaNValues(dthr_in_uti, dthr_out_uti):
    isNaN = False
    _aux = [str(v) for v in dthr_in_uti]
    for v in dthr_out_uti:
        _aux.append(str(v))
    for r in _aux:
        if r == 'nan':
            isNaN = True
    return isNaN
```

```python
duration = 0
estudo_comorb_prev_outras, cd_unidade = [], []
data_coleta, dthr_alta = [], []
hr_entrada_uti, hr_lib_mov_uti = [], []
new_csv_file = open('./DataSteps/STEP_6_Final_Dataset.csv', 'w+', encoding='utf8')
new_csv_file.write('\t'.join(df_sdn.columns.to_list())+'\tSEGUNDOS_NA_UTI\tDIAS_NA_UTI\n')

#for cd_a in list_unique_cd_atendimento:
for cd_a in [1403125.0]:
    isnan = False
    components = df_sdn.loc[df_sdn['CD_ATENDIMENTO']==cd_a].values.tolist()
    if len(components) == 1:
        aux_list = [str(value) for value in components[0]]
        try: 
            if np.isnan(components[0][121]) == True and np.isnan(components[0][122]) == True:
                new_csv_file.write('\t'.join(aux_list)+'\t'+'0.0'+'\t'+'0.0'+'\n')
                isnan = True
            else:
                isnan = False
        except:
            if str(components[0][121]) == 'nan' and str(components[0][122]) == 'nan':
                new_csv_file.write('\t'.join(aux_list)+'\t'+'0.0'+'\t'+'0.0'+'\n')
                isnan = True
            else:
                isnan = False
                
        if isnan == False:
            _, _, seconds = __getDifferenceInSeconds(components[0][121], components[0][122])
            days_uti = seconds/86400
            new_csv_file.write('\t'.join(aux_list)+'\t'+str(seconds)+'\t'+str(days_uti)+'\n')
            
    if len(components) > 1:
        for cpt in components:
            estudo_comorb_prev_outras.append(str(cpt[84]))
            data_coleta.append(cpt[10])
            dthr_alta.append(cpt[27])
            cd_unidade.append(str(cpt[115]))
            hr_entrada_uti.append(cpt[121])
            hr_lib_mov_uti.append(cpt[122])
        if __isNaNValues(hr_entrada_uti, hr_lib_mov_uti) == False:
            if __checkDatesColetaAlta(data_coleta, dthr_alta) == 'equal':
                for i in range(len(hr_entrada_uti)):
                    _, _, seconds = __getDifferenceInSeconds(hr_entrada_uti[i], hr_lib_mov_uti[i])
                    duration+=seconds
                    seconds = 0
            aux_list = __getListWthMultipleValues(components[0], estudo_comorb_prev_outras, cd_unidade, hr_entrada_uti, hr_lib_mov_uti)
            days_uti = duration/86400
            print(aux_list)
            new_csv_file.write('\t'.join(aux_list)+'\t'+str(duration)+'\t'+str(days_uti)+'\n')
        else:
            hr_entrada_uti = [str(v) for v in hr_entrada_uti]
            hr_lib_mov_uti = [str(v) for v in hr_lib_mov_uti]
            aux_list = __getListWthMultipleValues(components[0], estudo_comorb_prev_outras, cd_unidade, hr_entrada_uti, hr_lib_mov_uti)
            new_csv_file.write('\t'.join(aux_list)+'\t'+'0.0'+'\t'+'0.0'+'\n')
    estudo_comorb_prev_outras, cd_unidade = [], []
    data_coleta, dthr_alta = [], []
    hr_entrada_uti, hr_lib_mov_uti = [], []
    duration = 0
```
