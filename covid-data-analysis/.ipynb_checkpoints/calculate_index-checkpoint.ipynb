```python
!pip install seaborn

# Importing libs EDA
import pandas as pd
import numpy as np

# Importing libs VIZ
import matplotlib.pyplot as plt
import seaborn as sns

# Read csv file
df = pd.read_csv('./data//Dataset_DiasNaUTI.csv', delimiter='\t')
df.head()
```

```python
df['ESTUDO_CHARLSON']
```

```python
lista = df.columns.to_list()
lista_colunas = []
for i in lista:
    if i.startswith('ESTUDO_COMORB_PREV_'):
        lista_colunas.append(i)
        
    print(lista.index(i), i)
    

#print(lista_colunas)
```

```python
df_cii = df[["CD_PACIENTE", "IDADE", *lista_colunas]]
df_cii = df_cii.rename(columns=str.lower)
df_cii.idade = df_cii.idade.str.split(',').str[0].astype(int)
df_cii.head()
```

```python
# Missing Values of Column
def verificarMissing(coluna, df):
    missingC = df[coluna].isnull().sum()
    print("Valores Faltantes: ", missingC)
    return missingC

# Create a list of fields presented in column 
def listarUnicos(valor, df):
    ndf = df[~df[valor].isna()].copy()
    lista = ndf[valor].unique().tolist()
    print(valor+": ", lista)
    return lista

# Resume
def exibirResumo(campo, df):
    listarUnicos(campo, df)
    verificarMissing(campo, df)
    print('//')
    
    
```

```python
exibirResumo('ESTUDO_CHARLSON', df)
```

```python
for value in df_cii.columns:
    exibirResumo(value, df_cii)
```

```python
search = ' // ' or','
df_ciiF['count_outras']= df_cii['estudo_comorb_prev_outras'].astype(str).str.count('//|,')

#df_cii["count_outras"]

plt.figure(figsize=(20, 10))
sns.histplot(df_ciiF.count_outras, kde=True)
plt.title("Quantidade de comorbidades")
plt.show()
```

```python
charlson = df_ciiF[[*columns_cii]].sum(axis = 1, skipna = True) 
print(charlson)
```

```python
plt.figure(figsize=(20, 10))
sns.histplot(charlson, kde=True)
plt.title("Índice Charlson")
plt.show()
```

```python
idade_cii = []    

for idade in df_ciiF.idade:
    if (idade >= 80):
        #print('4')
        idade_cii.append(4)
    elif (idade >= 70 and idade <=79):
        #print('3')
        idade_cii.append(3)
    elif (idade >= 60 and idade <=69):
        #print('2')
        idade_cii.append(2) 
    elif (idade >= 50 and idade <=59):
        #print('1')
        idade_cii.append(1)
    else:        
        #print('0')
        idade_cii.append(0)
        
# Índice de comorbidade para idade
print(idade_cii)    
```

```python
df_ciiF['idade'] = idade_cii
df_ciiF['idade']
```

```python
charlson_idade = df_ciiF[['idade', *columns_cii]].sum(axis = 1, skipna = True) 
print(charlson_idade)
```

```python
plt.figure(figsize=(20, 10))
sns.histplot(charlson_idade, kde=True)
plt.title("Índice Charlson unindo com Idade")
plt.show()
```
