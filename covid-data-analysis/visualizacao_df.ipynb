```python
# Missing Values of Column
def verificarMissing(coluna):
    missingC = df[coluna].isnull().sum()
    print("Valores Faltantes: ", missingC)
    return missingC
```

```python
# Create a list of fields presented in column 
def listarUnicos(valor):
    ndf = df[~df[valor].isna()].copy()
    lista = ndf[valor].unique().tolist()
    print("Lista: ", lista)
    return lista
```

```python
# Create a Wordcloud with function
def print_image(valor):
    wc_d = WordCloud(use_tfidf=False, stopwords = PORTUGUESE_STOPWORDS)
    #print(valor)
    embed_code_d = wc_d.get_embed_code(text = valor,random_color = True,topn=100)
    return HTML(embed_code_d)
```

```python
missingMunicipioResidencia = verificarMissing('Município de Residência')
listaDeMunicipios = listarUnicos('Município de Residência')
```

```python
missingDataNascimento = verificarMissing('Data_Nascimento')
listaNascimento = listarUnicos('Data_Nascimento')
```

```python
from datetime import datetime
import datetime
import time
from dateutil.relativedelta import relativedelta

today = datetime.datetime.now() 
print(today)
print(type(today))
```

```python
ageList = []
for data in listaNascimento:
    if len(data) == 10:
        data = data.replace('/19/','/09/')
        # convert str to datetime
        dataNas = datetime.datetime.strptime(data, '%d/%m/%Y')
        rdata = relativedelta(today, dataNas)
        ageList.append(rdata.years)
        

dfAge = pd.DataFrame(ageList,columns=['Idade'])
dfAge.value_counts()
```

```python
plt.figure(figsize=(20,10))
# Age Class Per UsefulCount
dfAge.value_counts().nlargest(20).plot(kind='bar')
plt.title("Proporção de Idade")
# Descomentar para salvar imagem
# plt.savefig("idades.png")
plt.show()
```
